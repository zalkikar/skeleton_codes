### Linear Regression

```python
def feature_normalization(train, test):
    train_mins = np.amin(train,axis=0)
    train_maxs = np.amax(train,axis=0)
    train_normalized = (train-train_mins)/(train_maxs-train_mins)
    test_normalized = (test-train_mins)/(train_maxs-train_mins)
    return train_normalized, test_normalized
```

```python
def compute_square_loss(X, y, theta):
    loss = 0
    num_instances = X.shape[0]
    term = np.dot(X,theta)
    return np.dot(np.transpose(term-y),term-y)/num_instances
```

```python
def compute_square_loss_gradient(X, y, theta):
    num_instances= X.shape[0]
    term = np.dot(X,theta)
    grad = 2* (term-y).dot(X)/num_instances
    return grad
```

```python
def batch_grad_descent(X, y, alpha=0.1, num_step=1000, grad_check=False):
    num_instances, num_features = X.shape[0], X.shape[1] # create batches before you do this
    theta_hist = np.zeros((num_step+1, num_features)) #Initialize theta_hist
    loss_hist = np.zeros(num_step+1) #Initialize loss_hist
    theta = np.zeros(num_features) #Initialize theta
    
    if (grad_check == True)&(grad_checker(X,y,theta,epsilon=0.01,tolerance=1e-4) == False): return False
    for i in range(0,num_step+1):
        if (grad_check == True)&(not grad_checker(X,y,theta,epsilon=0.01,tolerance=1e-4)):return False
        if i > 0: theta-=(alpha*compute_square_loss_gradient(X,y,theta))
        loss_hist[i] = compute_square_loss(X,y,theta)
        theta_hist[i,:] = theta
    return theta_hist,loss_hist
```

### Ridge Regression

```python
def compute_regularized_square_loss_gradient(X, y, theta, lambda_reg):
    return 2 * np.mean((np.dot(X,theta)-y).reshape(-1,1)*X,axis=0) + 2*lambda_reg*theta
```

```python
def regularized_grad_descent(X, y, alpha=0.05, lambda_reg=10**-2, num_step=1000):
    num_instances, num_features = X.shape[0], X.shape[1]
    theta = np.zeros(num_features) #Initialize theta
    theta_hist = np.zeros((num_step+1, num_features)) #Initialize theta_hist
    loss_hist = np.zeros(num_step+1) #Initialize loss_hist
    
    for i in range(0,num_step+1):
        if i > 0: theta-=(alpha*compute_regularized_square_loss_gradient(X,y,theta,lambda_reg))
        loss_hist[i] = compute_square_loss(X,y,theta)
        theta_hist[i,:] = theta
    return theta_hist,loss_hist
```

The ideal intercept is fixed, and we know that the dot product of $B$ and our weights are equal to the ideal intercept (which is governed by the ideal decision boundary). Therefore, it is easy to see that as you increase to a large $B$, the affect is a decrease in the weights, forcing $\lambda \Theta$ to go down, reducing effective regularization on the bias term.

### Stochastic Gradient Descent

```python
def stochastic_grad_descent(X, y, alpha=0.01, lambda_reg=10**-2, num_epoch=1000):
    num_instances, num_features = X.shape[0], X.shape[1]
    theta = np.ones(num_features) #Initialize theta
    theta_hist = np.zeros((num_epoch, num_instances, num_features)) #Initialize theta_hist
    loss_hist = np.zeros((num_epoch, num_instances)) #Initialize loss_hist
    C=0.1
    sqrt = False
    is_string = False
    
    if str(alpha)=="1/sqrt(t)": 
        sqrt = True
        is_string = True
        
    elif str(alpha)=="1/t": 
        sqrt = False
        is_string = True
        
    else:
        alpha = float(alpha)
    
    for ind in range(num_epoch):
        shuffle = np.arange(num_instances)
        np.random.shuffle(shuffle)
        if is_string == True:
            if ind==0: 
                alpha = C
            elif ind>0 and sqrt == True:
                alpha = C/np.sqrt(ind)
            elif ind>0 and sqrt == False:
                alpha = C/ind
            
        for s in shuffle:
            theta -= 2*alpha*((np.dot(theta,X[s,:])-y[s])*X[s,:]+lambda_reg*theta)/num_instances
            loss_hist[ind,s] = compute_square_loss(X,y,theta)
            theta_hist[ind,s,:] = theta
            
    return theta_hist,loss_hist
```

It seems as though SGD converges worst with fixed step size, and it appears as though alpha = 0.1/sqrt(t) is the best. SGD can be better than regular GD because it takes many more steps than regular GD resulting in a model that is better.

### SVM with Pegasos

```python
def pegasos(X,y,lambda_reg,max_epoch):
    w = {}
    epoch = 0
    t = 1
    while epoch < max_epoch:
        epoch = epoch + 1
        for j in range(len(X)):
            t = t + 1
            eta = 1 / (t * lambda_reg)
            
            # always do this 
            increment(w, -eta * lambda_reg, w)
            
            # also do this if less than 1
            if (y[j] * dotProduct(w,X[j])) < 1:
                increment(w, eta * y[j], X[j])
    return w
```

```python
def pegasos_faster(X,y,lambda_reg,max_epoch):
    w = {}
    epoch = 0
    t = 1
    s = 1
    while epoch < max_epoch:
        epoch = epoch + 1
        for j in range(len(X)):
            t = t + 1
            eta = 1 / (t * lambda_reg)
            s = s + (-eta * lambda_reg * s)
            
            if (y[j] * dotProduct(w,X[j]) * s) < 1:
                increment(w, (1/s) * eta * y[j], X[j])
                
    w.update((key,s*value) for key,value in w.items())
    return w
```

```python
def evaluate(w,X,y):
    errors = 0
    for i in range(len(X)):
        predict = 0
        if dotProduct(w,X[i]) < 0:
            predict = -1
        else:
            predict = 1
            
        if predict != y[i]: 
            errors = errors + 1
            
    return errors / len(X)
```

```python
def pegasos_faster_earlystopping(X,y,lambda_reg,max_epoch):
    w = {}
    epoch = 0
    t = 1
    s = 1
    train_losses = []
    while epoch < max_epoch:
        epoch = epoch + 1
        for j in range(len(X)):
            t = t + 1
            eta = 1 / (t * lambda_reg)
            s = s + (-eta * lambda_reg * s)
            
            if (y[j] * dotProduct(w,X[j]) * s) < 1:
                increment(w, (1/s) * eta * y[j], X[j])
                
        train_losses.append(evaluate(w,X,y))
        if len(train_losses) > 5:
            stop = 0
            for l in train_losses[-5:]:
                if l > train_losses[-6]:
                    stop = stop + 1
            if stop == 5:
                print("Early stop at epoch {}".format(epoch))
                epoch = max_epoch + 1 # early stop
        
    w.update((key,s*value) for key,value in w.items())
    
    return w
```
